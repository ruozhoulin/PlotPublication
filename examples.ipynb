```python
from plot_publication import *
import numpy as np
```

```python
def set_label(ax):
    ax.set_xlabel('X label')
    ax.set_ylabel('Y label')

x = np.linspace(1, 10, 50)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.log(x)

Fig = FigurePublication(1, 1)
fig, ax = Fig.fig, Fig.ax
ax.plot(x, y1, label='sin')
ax.plot(x, y2, label='cos')
ax.plot(x, y3, label='tan')
legend(ax)
set_label(ax)
Fig.save('figures/1-1.svg')
```

```python
Fig = FigurePublication(1, 2)
fig, axs = Fig.fig, Fig.ax

names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]
axs[0].bar(names, values, label='Bar plot')
axs[1].scatter(names, values, label='Scatter plot')

for ax in axs:
    set_label(ax)
    legend(ax)

# fig.suptitle('Categorical Plotting')
Fig.update()
Fig.save('figures/1-2.svg')
```

```python
x = np.linspace(1, 10, 50)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.sinh(x)
y5 = np.cosh(x)
y6 = np.tanh(x)
y7 = np.exp(x)
y8 = np.log(x)
ys = [y1, y2, y3, y4, y5, y6, y7, y8]

Fig = FigurePublication(2, 2)
for i, ax in enumerate(Fig.ax.flatten()):
    ax.plot(x, ys[i],label=i)
    set_label(ax)
    legend(ax)
Fig.update()
Fig.save('figures/2-2.svg')
```

```python
Fig = FigurePublication(3, 2)
for i, ax in enumerate(Fig.ax.flatten()):
    ax.plot(x, ys[i], label=i)
    set_label(ax)
    legend(ax)
Fig.update()
Fig.save('figures/3-2.svg')
```

```python
Fig = FigurePublication(4, 2)
colors = get_default_color() # default color from matplotlib
for i, ax in enumerate(Fig.ax.flatten()):
    ax.plot(x, ys[i], color=colors[i], label=i)
    set_label(ax)
    legend(ax)
Fig.update()
Fig.save('figures/4-2.svg')
```
