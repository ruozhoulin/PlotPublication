```python
Fig = FigurePublication(1, 2)
fig, axes = Fig.fig, Fig.ax

names = ["group_a", "group_b", "group_c"]
values = [1, 10, 100]

ax = axes[0]
ax.cla()  # clear the current Axes state without closing it.
ax.bar(names, values, label="Bar plot")
enable_axes_legend(ax)

unitX = UnitTex.time("s")
ax.set_xlabel("Time (%s)" % unitX)
unitY = UnitTex.discharge('m', 's')
ax.set_ylabel("Flow rate (%s)" % unitY)

ax = axes[1]
ax.cla()  # clear the current Axes state without closing it.
ax.scatter(names, values, label="Scatter plot")
enable_axes_legend(ax)

unitX = UnitTex.time("d")
ax.set_xlabel("Time (%s)" % unitX)
unitY = UnitTex.area("mm")
ax.set_ylabel("Area (%s)" % unitY)

Fig.save(os.path.join("figures", "unit.png"))
```
