# Observation vs Prediction

```python
# plot
Fig = FigurePublication(1, 1)
fig, axes = Fig.fig, Fig.ax
ax: plt.Axes = axes

ax.clear()  # clear the current Axes state without closing it.

ax.set_box_aspect(1) # make a square box

# generate data for illustration
x = np.random.rand(100) * 2
y = x + np.random.rand(100)

# * draft plot x-y
ax.plot(x, y, "g", alpha=0.5, linestyle="", marker=".", markerfacecolor="none")

# * get maximum range of each ax
limits = [
    np.max([np.min([ax.get_xlim(), ax.get_ylim()]), 0]),  # min of both axes, but no smaller than 0
    np.max([ax.get_xlim(), ax.get_ylim()]),  # max of both axes
]
# adjust limits of x and y
ax.set_xlim(limits)
ax.set_ylim(limits)

# * plot y=x according to the limits of x and y
ax.plot(limits, limits, "k-", alpha=0.8, zorder=0)

set_tick_number_x(3, ax)
set_tick_number_y(3, ax)
enable_minor_locator(ax)
ticks_align_limits_x(ax)
ticks_align_limits_y(ax)

ax.set_xlabel("Estimated discharge (%s)" % UnitTex.discharge("m", "s"))
ax.set_ylabel("Observed discharge (%s)" % UnitTex.discharge("m", "s"))

# * add statistic metrics
nse = 0.99 # r2(np.array(x), np.array(y))
loss = 0.1 # mse(np.array(x), np.array(y))
text_metric = "$\mathrm{MSE} = %.4f$\n$\mathrm{R^2} =%.4f$" % (loss, nse)

Fig.corner_annotate(ax, text_metric)

# fig.tight_layout()
```

# Color bar

# Method 1

```python
Fig = FigurePublication(1, 1)
fig, ax = Fig.fig, Fig.ax

np.random.seed(999)
matrix = np.random.rand(3,3)

# control range of color
vmin = np.min(matrix)
vmax = np.max(matrix)
dv = vmax - vmin
vmin -= 0.3 * dv
vmax += 0.01 * dv

# https://matplotlib.org/stable/gallery/images_contours_and_fields/interpolation_methods.html
pos = ax.imshow(matrix, interpolation="bilinear", cmap=matplotlib.colormaps["Reds"], vmin=vmin, vmax=vmax)
ax.set_xlabel("X label")
ax.set_ylabel("Y label")

# * set color bar
# https://matplotlib.org/stable/gallery/color/colorbar_basics.html
# clb = fig.colorbar(pos, ax=ax, location="right", fraction=1, pad=0.08)
clb = fig.colorbar(pos, ax=ax, location="right", fraction=0.05, pad=0.08)

tick_locator = MaxNLocator(nbins=9)
clb.locator = tick_locator
clb.update_ticks()

clb.ax.set_title("Title")
clb.set_label("Label")
```

## Method 2

Looks better but more complicated.

```python
Fig = FigurePublication(1, 1)
fig, axes = Fig.fig, Fig.ax

fig.clear()


w_ratios = [20, 1]
h_ratios = [1]

row_count = 1
col_count = 2
# * specify grids for creating axes with different size
gs = fig.add_gridspec(nrows=row_count, ncols=col_count, width_ratios=w_ratios, height_ratios=h_ratios)

# * create each axes
ax1 = fig.add_subplot(gs[0, 0])
ax1.set_box_aspect(1)  # image usually displayed in square
ax1.set_xlabel("X label")
ax1.set_ylabel("Y label")

ax_cb = fig.add_subplot(gs[0, 1])

# set proper figure width, if width is too small, the color bar will much longer than the sub-figure created by imshow(),
# because the sub-figure will be compressed to achieve aspect ratio equal 1.0
Fig.stretch_figure_width(1.05)
```
