# Observation vs Prediction

```python
# plot
Fig = FigurePublication(1, 1)
fig, axes = Fig.fig, Fig.ax
ax = axes

ax.cla()  # clear the current Axes state without closing it.

# generate data for illustration
x = np.random.rand(100) * 2
y = x + np.random.rand(100)

# * draft plot x-y
ax.plot(x, y, "g", alpha=0.5, linestyle="", marker=".", markerfacecolor="none")

# * get maximum range of each ax
limits = [
    np.max([np.min([ax.get_xlim(), ax.get_ylim()]), 0]),  # min of both axes, but no smaller than 0
    np.max([ax.get_xlim(), ax.get_ylim()]),  # max of both axes
]
# adjust limits of x and y
ax.set_xlim(limits)
ax.set_ylim(limits)

# * plot y=x according to the limits of x and y
ax.plot(limits, limits, "k-", alpha=0.8, zorder=0)

# set_tick_number_x(7, ax)
# set_tick_number_y(7, ax)

ax.set_xlabel("Estimated discharge (%s)" % UnitTex.discharge("m", "s"))
ax.set_ylabel("Observed discharge (%s)" % UnitTex.discharge("m", "s"))

# * add statistic metrics
nse = 0.99 # r2(np.array(x), np.array(y))
loss = 0.1 # mse(np.array(x), np.array(y))
text_metric = "$\mathrm{MSE} = %.4f$\n$\mathrm{R^2} =%.4f$" % (loss, nse)

cornor_annotate(ax, text_metric)

fig.tight_layout()
```

# Color bar

```python
Fig = FigurePublication(1, 1)
fig, ax = Fig.fig, Fig.ax

np.random.seed(999)
matrix = np.random.rand(3,3)

# control range of color
vmin = np.min(matrix)
vmax = np.max(matrix)
dv = vmax - vmin
vmin -= 0.3 * dv
vmax += 0.01 * dv

# https://matplotlib.org/stable/gallery/images_contours_and_fields/interpolation_methods.html
pos = ax.imshow(matrix, interpolation="bilinear", cmap=matplotlib.colormaps["Reds"], vmin=vmin, vmax=vmax)
ax.set_xlabel("X label")
ax.set_ylabel("Y label")

# * set color bar
# https://matplotlib.org/stable/gallery/color/colorbar_basics.html
clb = fig.colorbar(pos, ax=ax, location="right", fraction=0.03, pad=0.08)

tick_locator = MaxNLocator(nbins=9)
clb.locator = tick_locator
clb.update_ticks()

clb.ax.set_title("Title")
clb.set_label("Label")
```
